```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from textblob import TextBlob, Word
%matplotlib inline
import string
```

```python
jobs = pd.read_csv('./datasets/jobs_clean.csv',encoding = 'utf-8')
jobs.drop([jobs.columns[0]], axis=1, inplace=True)
```

```python
jobs.head()
```

```python
jobs.company = jobs.company.apply(lambda x: x.replace('\n',' '))
jobs.company = jobs.company.apply(lambda x: x.lower().strip())
jobs['company'] = jobs['company'].apply(lambda x:''.join([i for i in x if i not in string.punctuation]))
```

```python
jobs.head()
```

```python
dummy = pd.get_dummies(jobs, columns =['state'], drop_first=True, prefix=['state'])
jobs.drop('state',axis=1,inplace=True)
jobs = pd.concat([jobs, dummy])
```

```python
dummy = pd.get_dummies(jobs, columns =['company'], drop_first=True, prefix=['coy'])
jobs.drop('company',axis=1,inplace=True)
jobs = pd.concat([jobs, dummy])
```

```python
jobs.shape
```

```python
jobs.drop(jobs[jobs['salary_'] == 'None'].index,inplace=True)
```

```python
stop = stopwords.words('english')
jobs['details'] = jobs['details'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
jobs.head()
```

```python
vectorizer = CountVectorizer(ngram_range=(1, 1),min_df=5)
X=vectorizer.fit_transform(list(jobs['details']))
counts=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
```
