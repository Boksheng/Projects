```python
from bs4 import BeautifulSoup
import urllib

from selenium import webdriver


import os
import pandas as pd
import numpy as np
import pickle
import re
import time

import string
```

```python
chromedriver = './chromedriver/chromedriver'
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(executable_path = chromedriver)
```

```python
urls = []

for pageno in range(0,100):
    print(pageno)
    url = 'https://uk.reuters.com/news/archive/oilRpt?view=page&page='+str(pageno)+'&pageSize=10'
    driver.get(url)
    time.sleep(3 + 8*np.random.uniform())

    soup = BeautifulSoup(driver.page_source, 'lxml')
    for div in soup.find_all('div', attrs={'class': 'story-content'}):
        url = 'https://uk.reuters.com{}'.format((div.findAll('a')[0].attrs['href']))
        urls.append(url)
```

```python
driver.close()
```

```python
len(urls)
```

```python
news_text = []
title = []

for i,url in enumerate(urls[0:500]):
    if i == len(urls):
        print('Last')
    else:
        print(i)
    driver.get(url)
    time.sleep(2+5*np.random.uniform())
    
    try:
        soup = BeautifulSoup(driver.page_source,'lxml')
        for div in (soup.find_all('h1', attrs={'class': 'headline_2zdFM'})):
            news = div.text.replace('\n', '')
            title.append(news)
    except AttributeError:
        title.append('NA')   
    
    try:
        soup = BeautifulSoup(driver.page_source,'lxml')
        for div in (soup.find_all('div', attrs={'class': 'body_1gnLA'})):
            news = div.text.replace('\n', '')
            news_text.append(news)
    except AttributeError:
        news_text.append('NA')
```

```python
data_dict_1 = {'title': title, 'news_text': news_text}
df_1 = pd.DataFrame(data_dict_1)
df_1.head()
```
