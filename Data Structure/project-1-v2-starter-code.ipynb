## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
# Function to determine the aggregate of player's pokemons' power
def player_power(players_dict,pokedex_dict,player_id):
    power_total = 0
    # Iterate through player's pokemon's list
    for pokemon_id in players_dict[player_id]['player_pokemon']:
        power = 0
        # Sum of pokemon's attack, defense, s_attack and s_defense from pokedex dictionary
        power = (pokedex_dict[pokemon_id]['attack']
                + pokedex_dict[pokemon_id]['defense']
                + pokedex_dict[pokemon_id]['special_attack']
                + pokedex_dict[pokemon_id]['special_defense'])
        power_total += power
    # Print out Power of Player
    print('{player_name}\'s Power is {player_power}.'.format(player_name = players_dict[player_id]['player_name'],
                                                             player_power = power_total))
````

````python
# Iterate through each element of the above list and splitting by ','
raw_pd_list_01 = []
for i in raw_pd_list_00:
    raw_pd_list_01.append(i.split(','))
    
# printing out 1st and 2nd element of the above list
print(raw_pd_list_01[0:2])
````

````python
# Iterate through each element of the above list and remove "
pd_rmv_quotes = []
for i in raw_pd_list_01:
    pd_elmts_rmv_quotes = []
    for j in i:
        pd_elmts_rmv_quotes.append(j.replace('"',''))
    pd_rmv_quotes.append(pd_elmts_rmv_quotes)
    
# printing out 1st three elements of the above list
print(pd_rmv_quotes[0:10])
````

````python
# Iterate through each element of the above list and replace "" with -1, float numeric values
pd_rplemp_float = []
for i in pd_rmv_quotes:
    pd_elmts_rplemp_float = []
    for j in i:
        # Execute if cell is empty
        if j == '':
            pd_elmts_rplemp_float.append(float(-1))
        else:
            # Execute if cell value is numeric
            try:
                pd_elmts_rplemp_float.append(float(j))
            except:
                pd_elmts_rplemp_float.append(j.strip('\n'))
    pd_rplemp_float.append(pd_elmts_rplemp_float)
    
# printing out 1st three elements of the above list
print(pd_rplemp_float[0:3])
````

````python
# function to append items from pokedex if fit string conditions
def str_filt(d,k,v):
    temp = {}
    for i,j in d.items():
        if j[k] == v:
            temp[i] = d[i]
    return temp
````

````python
# function to append items from pokedex if fit numeric conditions
def num_filt(d,k,v):
    temp = {}
    for i,j in d.items():
        if j[k] >= v:
            temp[i] = d[i]
    return temp
````

````python
# function to append items into one dictionary
def filt_dex(d1,d2):
    for i in d2:
        if type(d2[i]) == int or type(d2[i]) == float:
            d1 = num_filt(d1,i,d2[i])
        else:
            d1 = str_filt(d1,i,d2[i])
    return(d1)
````
